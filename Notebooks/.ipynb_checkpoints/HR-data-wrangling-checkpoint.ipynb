### 2.6.1 Number Of Missing Values By Column<a id='2.6.1_Number_Of_Missing_Values_By_Column'></a>


```python
#Count (using `.sum()`) the number of missing values (`.isnull()`) in each column of 
#ski_data as well as the percentages (using `.mean()` instead of `.sum()`).
#Order them (increasing or decreasing) using sort_values
#Call `pd.concat` to present these in a single table (DataFrame) with the helpful column names 'count' and '%'
missing = pd.concat([hr_data.isnull().sum(), 100 * hr_data.mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by=['count'], ascending = False)
```

There is no column with missing value. Some of the columns have value '0' in multiple rows like, this could be part of data.
